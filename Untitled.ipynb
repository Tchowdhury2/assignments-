```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import solve_ivp

# Constants
temp = 298  # Kelvin
pres = 1e5  # Pascals
Rgas = 8.314  # J/(mol-K)
avo = 6.022e23  # molecules/mol
PHOx = 0.1  # pptv/s

# Rate constants (units: cm/(molecule-s), except where noted)
k1 = 26.3e-12
k2 = 7.7e-12
k3 = 8.1e-12
k4 = 1.1e-11
k5 = 2.9e-12
k6 = 5.2e-12
k7 = 0.015  # s
k8 = 1.9e-14

# Unit conversion: air density in molecules/cm
airden = (pres * avo) / (Rgas * temp * 1e6)  # molecules/cm

# Function for chemical reactions
def myfun(t, u):
    f = np.zeros(7)
    f[0] = -k2 * u[3] * u[0] + k7 * u[1] - k8 * u[5] * u[0] - k3 * u[4] * u[0]
    f[1] = k2 * u[3] * u[0] + k3 * u[4] * u[0] + k8 * u[5] * u[0] - k7 * u[1] - k4 * u[2] * u[1]
    f[2] = -k1 * u[6] * u[2] + k3 * u[4] * u[0] - k4 * u[2] * u[1] + CPHOx
    f[3] = k1 * u[6] * u[2] - k2 * u[3] * u[0] - k6 * u[3] * u[4]
    f[4] = k2 * u[3] * u[0] - k3 * u[4] * u[0] - 2 * k5 * u[4]**2 - k6 * u[3] * u[4]
    f[5] = k7 * u[1] - k8 * u[5] * u[0]
    f[6] = -k1 * u[6] * u[2]
    return f

# Function to calculate maximum O3 for given NOx and RH
def calc_max_o3(nox_ppb, rh_ppb):
    # Convert NOx and RH mixing ratios to molecules/cm
    Cnox = nox_ppb * 1e-9 * airden
    Crh = rh_ppb * 1e-9 * airden
    global CPHOx
    CPHOx = PHOx * 1e-12 * airden  # molecules/(cm-s)

    # Initial conditions
    u0 = np.zeros(7)
    u0[0] = Cnox * (2 / 3)  # NO
    u0[1] = Cnox * (1 / 3)  # NO2
    u0[2] = 0  # OH
    u0[3] = 0  # RO2
    u0[4] = 0  # HO2
    u0[5] = 0  # O3
    u0[6] = Crh  # RH (hydrocarbon)

    # Solve ODE
    t0 = 0
    tmax = 96 * 3600  # 96 hours
    sol = solve_ivp(myfun, [t0, tmax], u0, method='LSODA', t_eval=np.linspace(t0, tmax, 1000))

    # Return maximum O3 in ppb
    return np.max(sol.y[5] * (1e9 / airden))

# Create grid of NOx and RH values
nox_vals = np.linspace(0, 200, 50)  # Range of NOx (ppb)
rh_vals = np.linspace(0, 2000, 50)  # Range of RH (VOC in ppbC)
O3_max = np.zeros((len(rh_vals), len(nox_vals)))

# Nested loop to calculate O3 for each NOx and RH combination
for i, rh in enumerate(rh_vals):
    for j, nox in enumerate(nox_vals):
        O3_max[i, j] = calc_max_o3(nox, rh)
        print(f"Calculating O3: NOx={nox} ppb, RH={rh} ppb -> Max O3={O3_max[i, j]:.2f} ppb")

# Contour plot for Ozone Isopleth
plt.figure(figsize=(10, 6))
levels = np.arange(0, 200, 10)  # Contour levels every 10 ppb
contour = plt.contourf(nox_vals, rh_vals, O3_max, levels=levels, cmap='viridis')
plt.colorbar(contour, label='Maximum O3 (ppb)')
plt.xlabel('Initial NOx (ppb)')
plt.ylabel('Initial VOC (ppbC)')
plt.title('Ozone Isopleth Plot')
plt.show()
```
