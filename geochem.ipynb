```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
lambda_sm = 6.54e-12  # decay constant for 147Sm in yr^-1
initial_time = 4.567e9  # time in Ga, converted to years (4.567 Ga)
current_time = 0  # present day

# Initial values for the chondrite
initial_143Nd_144Nd_chondrite = 0.506756
initial_147Sm_144Nd_chondrite = 0.194

# Values for the gabbro today
current_143Nd_144Nd_gabbro = 0.511877
current_147Sm_144Nd_gabbro = 0.113

# Time range (from 4.567 Ga to today)
time_range = np.linspace(0, 4.567e9, 100)  # time in years

# Evolution of 143Nd/144Nd in chondrite due to radioactive decay
nd_chondrite = initial_143Nd_144Nd_chondrite + initial_147Sm_144Nd_chondrite * (
    np.exp(lambda_sm * time_range) - 1)

# Evolution of 143Nd/144Nd in gabbro (backtracked)
nd_gabbro = current_143Nd_144Nd_gabbro - current_147Sm_144Nd_gabbro * (
    np.exp(lambda_sm * (4.567e9 - time_range)) - 1)

# Plotting the results
plt.figure(figsize=(8, 6))
plt.plot(time_range / 1e9, nd_chondrite, label='Chondrite Evolution', color='blue')
plt.plot(time_range / 1e9, nd_gabbro, label='Gabbro Evolution', color='red')
plt.xlabel('Time (Ga)')
plt.ylabel('143Nd/144Nd')
plt.title('Evolution of 143Nd/144Nd with Time')
plt.legend()
plt.grid(True)
plt.show()
```
